### Task 3: Break down number of sets by year

```python
clean_df = merged[~merged['set_num'].isnull()]
clean_df['count'] = 1

clean_df.head()
```

```python

clean_df.groupby(['year']).sum().reset_index()[['year', 'count']]
```

```python
sets_per_year = clean_df.groupby(['year']).sum().reset_index()[['year', 'count']]
sets_per_year.head()
```

```python
sets_per_year.head(50)
```

```python
for index, row in sets_per_year.iterrows():
    print(row['year'], row['count'])
```
